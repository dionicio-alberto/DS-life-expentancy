# 📌 Distribución de Poisson en Python

La **distribución de Poisson** modela el número de eventos que ocurren en un intervalo de tiempo o espacio cuando estos eventos son independientes y ocurren con una tasa promedio constante \( \lambda \).

## 📌 **Fórmula Matemática**
La función de probabilidad de la distribución de Poisson está dada por:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

donde:
- $ k $ es el número de eventos que queremos calcular.
- $ \lambda $ es la tasa promedio de ocurrencia.
- $ e $ es la base del logaritmo natural.

## 🎯 **Ejemplo en Python**
Supongamos que en una tienda llegan en promedio **5 clientes por minuto** $( \lambda = 5 )$. Queremos calcular la probabilidad de que lleguen exactamente **3 clientes** en un minuto.

```python
from scipy.stats import poisson

# Parámetro lambda (tasa promedio de eventos por unidad de tiempo)
lambda_value = 5

# Calcular la probabilidad de que lleguen exactamente 3 clientes
k = 3
probabilidad = poisson.pmf(k, lambda_value)

print(f"La probabilidad de que lleguen exactamente {k} clientes es: {probabilidad:.4f}")

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetro lambda
lambda_value = 5

# Valores posibles de k
k_values = np.arange(0, 16)
probabilidades = poisson.pmf(k_values, lambda_value)

# Crear gráfico con Seaborn
plt.figure(figsize=(10, 5))
sns.barplot(x=k_values, y=probabilidades, color="royalblue")

# Línea vertical en la media
plt.axvline(lambda_value, color='red', linestyle='--', label=f'Media: {lambda_value}')

# Personalización del gráfico
plt.title("Distribución de Poisson (λ = 5)", fontsize=14)
plt.xlabel("Número de eventos (k)", fontsize=12)
plt.ylabel("Probabilidad", fontsize=12)
plt.legend()
plt.show()
````

📌 **Interpretación**

- La barra más alta indica el número de eventos más probable en un minuto.
- La línea roja punteada representa la media $E(X) = \lambda = 5 $
- La probabilidad disminuye para valores de $k$ mucho menores o mayores a la media.

## 🎯 **Ejemplo en Python: Probabilidad Acumulada**

Ahora queremos calcular la probabilidad de que lleguen como máximo 3 clientes en un minuto, es decir:

$$P(X≤3)=P(X=0)+P(X=1)+P(X=2)+P(X=3)$$



````python
# Calcular la probabilidad acumulativa P(X ≤ 3)
k = 3
probabilidad_acumulada = poisson.cdf(k, lambda_value)

print(f"La probabilidad de que lleguen como máximo {k} clientes es: {probabilidad_acumulada:.4f}")
````

### 📊 **Visualización de la Función Acumulativa (CDF)**
Ahora graficamos la CDF para ver cómo la probabilidad se acumula conforme aumenta $k$

````python
# Calcular valores de la CDF
cdf_values = poisson.cdf(k_values, lambda_value)

# Crear gráfico con Seaborn
plt.figure(figsize=(10, 5))
sns.lineplot(x=k_values, y=cdf_values, marker="o", color="darkorange")

# Línea horizontal en P(X ≤ 3)
plt.axhline(probabilidad_acumulada, color='green', linestyle='--', label=f'P(X ≤ {k}) = {probabilidad_acumulada:.3f}')
plt.axvline(k, color='green', linestyle='--')

# Personalización del gráfico
plt.title("Distribución de Poisson - Función de Distribución Acumulativa (CDF)", fontsize=14)
plt.xlabel("Número de eventos (k)", fontsize=12)
plt.ylabel("Probabilidad Acumulada", fontsize=12)
plt.legend()
plt.show()
````

### 📌 **Interpretación**

1. La PMF muestra la probabilidad exacta de que ocurran  $k$ eventos.
2. La CDF muestra la probabilidad acumulada de que ocurran como máximo $k$ eventos.
3. En este caso, la probabilidad de que lleguen 3 clientes o menos es aproximadamente 26.3%.
