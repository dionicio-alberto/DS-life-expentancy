# üìå Distribuci√≥n de Poisson en Python

La **distribuci√≥n de Poisson** modela el n√∫mero de eventos que ocurren en un intervalo de tiempo o espacio cuando estos eventos son independientes y ocurren con una tasa promedio constante \( \lambda \).

## üìå **F√≥rmula Matem√°tica**
La funci√≥n de probabilidad de la distribuci√≥n de Poisson est√° dada por:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

donde:
- $ k $ es el n√∫mero de eventos que queremos calcular.
- $ \lambda $ es la tasa promedio de ocurrencia.
- $ e $ es la base del logaritmo natural.

## üéØ **Ejemplo en Python**
Supongamos que en una tienda llegan en promedio **5 clientes por minuto** $( \lambda = 5 )$. Queremos calcular la probabilidad de que lleguen exactamente **3 clientes** en un minuto.

```python
from scipy.stats import poisson

# Par√°metro lambda (tasa promedio de eventos por unidad de tiempo)
lambda_value = 5

# Calcular la probabilidad de que lleguen exactamente 3 clientes
k = 3
probabilidad = poisson.pmf(k, lambda_value)

print(f"La probabilidad de que lleguen exactamente {k} clientes es: {probabilidad:.4f}")

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Par√°metro lambda
lambda_value = 5

# Valores posibles de k
k_values = np.arange(0, 16)
probabilidades = poisson.pmf(k_values, lambda_value)

# Crear gr√°fico con Seaborn
plt.figure(figsize=(10, 5))
sns.barplot(x=k_values, y=probabilidades, color="royalblue")

# L√≠nea vertical en la media
plt.axvline(lambda_value, color='red', linestyle='--', label=f'Media: {lambda_value}')

# Personalizaci√≥n del gr√°fico
plt.title("Distribuci√≥n de Poisson (Œª = 5)", fontsize=14)
plt.xlabel("N√∫mero de eventos (k)", fontsize=12)
plt.ylabel("Probabilidad", fontsize=12)
plt.legend()
plt.show()
````

üìå **Interpretaci√≥n**

- La barra m√°s alta indica el n√∫mero de eventos m√°s probable en un minuto.
- La l√≠nea roja punteada representa la media $E(X) = \lambda = 5 $
- La probabilidad disminuye para valores de $k$ mucho menores o mayores a la media.

## üéØ **Ejemplo en Python: Probabilidad Acumulada**

Ahora queremos calcular la probabilidad de que lleguen como m√°ximo 3 clientes en un minuto, es decir:

$$P(X‚â§3)=P(X=0)+P(X=1)+P(X=2)+P(X=3)$$



````python
# Calcular la probabilidad acumulativa P(X ‚â§ 3)
k = 3
probabilidad_acumulada = poisson.cdf(k, lambda_value)

print(f"La probabilidad de que lleguen como m√°ximo {k} clientes es: {probabilidad_acumulada:.4f}")
````

### üìä **Visualizaci√≥n de la Funci√≥n Acumulativa (CDF)**
Ahora graficamos la CDF para ver c√≥mo la probabilidad se acumula conforme aumenta $k$

````python
# Calcular valores de la CDF
cdf_values = poisson.cdf(k_values, lambda_value)

# Crear gr√°fico con Seaborn
plt.figure(figsize=(10, 5))
sns.lineplot(x=k_values, y=cdf_values, marker="o", color="darkorange")

# L√≠nea horizontal en P(X ‚â§ 3)
plt.axhline(probabilidad_acumulada, color='green', linestyle='--', label=f'P(X ‚â§ {k}) = {probabilidad_acumulada:.3f}')
plt.axvline(k, color='green', linestyle='--')

# Personalizaci√≥n del gr√°fico
plt.title("Distribuci√≥n de Poisson - Funci√≥n de Distribuci√≥n Acumulativa (CDF)", fontsize=14)
plt.xlabel("N√∫mero de eventos (k)", fontsize=12)
plt.ylabel("Probabilidad Acumulada", fontsize=12)
plt.legend()
plt.show()
````

### üìå **Interpretaci√≥n**

1. La PMF muestra la probabilidad exacta de que ocurran  $k$ eventos.
2. La CDF muestra la probabilidad acumulada de que ocurran como m√°ximo $k$ eventos.
3. En este caso, la probabilidad de que lleguen 3 clientes o menos es aproximadamente 26.3%.
